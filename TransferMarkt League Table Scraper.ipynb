```python
import requests
import re
from bs4 import BeautifulSoup
import pandas
```

```python
# Introducing ourselves
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36"
}
```

```python
# get the request based on the url
years = range(2020, 2025)
url = f"https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1/plus/?saison_id={years[-1]}"
response = requests.get(url, headers = headers)
```

```python
response.status_code # 200 means successful
```

```python
# get the html code
soup = BeautifulSoup(response.content, "html.parser") # response.content is the code html parses formats it
response.content
```

```python
# Select an html element. It remains a soup object
teams = soup.find_all('td', class_ = 'hauptlink no-border-links') # gives the text inside
```

```python
type(teams) # ResultSet object
```

```python
for td in teams:
    print(td.text)
```

```python
 team_elements = soup.find_all('a', class_='vereinprofil_tooltip')

# Extract and print the team names
team_names = [team.get_text(strip=True) for team in team_elements]
    
    # Print the team names
for name in team_names:
    print(name)
```
